```python
# making all val in crore

def convert_price(x):
    
    if type(x) == float:
        return x
    
    else:
        if x[1] == 'Lac':
            return round( float(x[0]) / 100 , 2 )
        else:
            return round( float(x[0]) , 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(convert_price)
```

```python
df.head(2)
```

```python
# price_per_sqft
df['price_per_sqft'].value_counts() # checking if anything is abnormal or  not
```

```python
# split on / , take 0th element rs.-->'' and ,--->'' , strip right left convert float bcoz none val
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
