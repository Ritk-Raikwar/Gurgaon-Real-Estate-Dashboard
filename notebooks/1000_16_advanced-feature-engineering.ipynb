```python
import numpy as np ; import pandas as pd
df = pd.read_csv('gurgaon_properties_post_feature_selection_v2.csv')
```

```python
df.head()
```

```python
# average price we calling it sector 
sp = df.groupby('sector')['price'].median() # price in crore denoted
sp
```

```python
sb = df.groupby('sector')['built_up_area'].median() # area in sq ft
sb
```

```python
# average price per sector
sector_price_map = sp/sb
sector_price_map
```

```python
sector_price_map['sector 28']
```

```python
df['sector'] # sector 28 will get mapped by the corresponding value sector_price_map ind sector val average price per sector
```

```python
df['sector'].map(sector_price_map)
```

```python
df['sector_avg_price']=df['sector'].map(sector_price_map)
```

```python
df.head()
```

```python
# another feature  = built_up_area * sector_avg_price
df['area_x_sector_avg_price'] = df['built_up_area'] * df['sector_avg_price']
```

```python
df['area_by_room'] = df['built_up_area'] / ( df['bedRoom'] + 1  )
```

```python
df['bed_bath_ratio'] = df['bedRoom'] / ( df['bathroom'] + 1 )
```

```python
df.head()
```

```python
df.to_csv('advanced-feature-dataset.csv',index=False)
```
